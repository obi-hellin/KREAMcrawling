# 1. 잘못 크롤링 된 행 제거하기


```python
dataI = data.dropna(subset = ['50'])
delist = dataI.index
data = data.drop(delist)
data = data.iloc[:,:58]
# data = data.drop(['50', '51', '52'], axis = 'columns')
data = data.reset_index(drop = True)
data.shape
```

```python
delindex = []
for i in range(data.shape[0]) :
    if ')' in str(data.iloc[i,8:].values.tolist()) :
        delindex.append(data.index[i])
    if '/' in str(data.iloc[i,8:].values.tolist()) :
        delindex.append(data.index[i])
    if any(j.isupper() for j in str(data.iloc[i,8:].values.tolist())) :
        delindex.append(data.index[i])
delindex = set(delindex)
delindex = list(delindex) # 중복 제거
data = data.drop(delindex)
data = data.reset_index(drop = True)
data.shape
# dataI.head()
```

# I. 가격 전처리. 출시 가격을 숫자로 바꾸기

```python
def chage(x):
    tmp = '0'
    if x=='-':
        pass
    elif x.find('약') > 0: # '약'이 있는것 체크
        idx = x.find('약')
        tmp = x[idx+1:-2]  # '원' 제거
    else:
        tmp =  x[:-1] # '원' 제거
    return tmp.replace(',','')
    

price_df = df['Release Price'].apply(chage)
price_df
```

# II. 출시 가격으로 거래 내역 결측치 채우기

```python
tmp2 = df.iloc[:,8:]
tmp2
```

```python
buffer = pd.DataFrame()
for i in range(tmp2.shape[0]):
    a = tmp2.iloc[ i:i+1, :]
    a.fillna( int(price_df.iloc[i].strip()), inplace=True )
    tmp2.iloc[ i:i+1, :] = a
    #print( tmp2)
    #break
tmp2
```

```python
testdf = tmp2.astype('str')
def changex(x) :
    x1 = '0'
    if x[-1] == '원' :
        x1 = x[:-1].replace(',',"")
    else :
        x1 = x
    return x1
for i in range(testdf.shape[1]) :
    testdf.iloc[:,i] = testdf.iloc[:,i].apply(changex)
testdf
```

```python
testdf = testdf.astype('int')
```

```python
price_df = price_df.astype('int')
```

# III. 신발 정보 데이터와 거래내역 데이터 합치기

```python
Newdf = Newdf.reset_index(drop = True)
price_df = price_df.reset_index(drop = True)
tmp2 = testdf.reset_index(drop = True)

KREAMDATA = pd.concat([Newdf, price_df, tmp2], axis = 1)
KREAMDATA = KREAMDATA.drop('Release Price', axis = 'columns')
KREAMDATA
```
